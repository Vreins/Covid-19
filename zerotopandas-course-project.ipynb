## Data Preparation and Cleaning

**TODO** - In this section the data would be cleaned. Columns with a lot of empty spaces will be removed and important columns with empty spaces will be removed by the rows



```python
import pandas as pd 
import numpy as np 
```

```python
covid_df=pd.read_csv('corona-virus-report/worldometer_data.csv')
```

```python
covid_df
```

```python
covid_df.info()
```

```python
covid_df.describe()
```

This is the total covid status of all countries from 209 countries from january 2020 to july 2020

```python
covid_df.drop(columns=['NewDeaths','NewCases','NewRecovered'], inplace=True)
```

```python
covid_df["Continent"].value_counts()
```

```python
covid_df.drop(covid_df[covid_df.TotalDeaths.isna()].index,inplace=True)
covid_df.drop(covid_df[covid_df.TotalTests.isna()].index,inplace=True)
covid_df.info()
```

## Exploratory Analysis and Visualization

**TODO** - The total death of the data cleaned would be calculated as well as the mean total death to ascertain later if the total death in each country was more than the mean death
Important relationship will be tabulated using pandas and visualized with matplotlib


Let's begin by importing`matplotlib.pyplot` and `seaborn`.

```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (9, 5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

```python
covid_df.describe()
```

**TODO** - The deaths due to covid would be calculated below as well as the positive rate in each country

```python
covid_df["death_rate"]=covid_df["TotalDeaths"]/covid_df["TotalCases"]*100
countries_death_rate=covid_df.sort_values("Population",ascending=False)
countries_death_rate.describe()
covid_df["positive_rate"]=covid_df["TotalCases"]/covid_df["TotalTests"]*100
covid_df.drop(covid_df[covid_df["positive_rate"]>100].index, inplace=True
)
```

```python
highest_death_rate=covid_df.sort_values("death_rate",ascending=False).head(15)
sns.barplot(x="Country/Region", y="death_rate", data=highest_death_rate)
plt.xticks(rotation=70);
plt.title("Countries with the highest death rates");
```

```python
lowest_death_rate=covid_df.sort_values("death_rate").head(15)
sns.barplot(x="Country/Region", y="death_rate", data=lowest_death_rate)
plt.title("Countries with the lowest death rates")
plt.xticks(rotation=70);
```

#### The graph for 15 countries with the highest death rate and 15 countries with the lowest death rates was plotted as shown above

```python
top_countries=countries_death_rate.head(15)
sns.barplot(x="Country/Region",y="death_rate", data=top_countries)
plt.title("Death rate of most populated countries");
plt.xticks(rotation=75);
```

```python
bottom_countries=countries_death_rate.tail(15)
sns.barplot(x="Country/Region",y="death_rate", data=bottom_countries);
plt.title("Death rate of least populated countries")
plt.xticks(rotation=75);
```

##### From both barplots above, we plotted two barplots to decipher if the death rate due to covid 19 could be as a result of the population. So I plotted the death rate of the highest populated countries as well as the death rates of the lowest populated countries in the dataframe. There was no marked difference between the death rate of least populated countries and most populated countries. That is not to say that the total death wouldn't have a marked difference

#### Q1: TODO - In the former section we saw that Europe has the highest death rates amongst the other continents, so the question is which countries in Europe has more death rate than the continental death rate and hence increasing the continental death rate

```python
european_countries=covid_df[covid_df.Continent=="Europe"]
european_countries.nunique()
```

```python
high_death_rate_countries=european_countries[european_countries.death_rate>european_countries.death_rate.mean()].sort_values("death_rate", ascending=False)
```

```python
sns.barplot(x=high_death_rate_countries["Country/Region"],y=high_death_rate_countries["death_rate"])
plt.xticks(rotation=70);
```

##### A barplot was plotted to decipher the countries with higher death rates than the continental death rate of Europe and it shows that France which was the country with the highest death was one of them followed by the Uk and 13 other countries

#### Q2: TODO - In the previous section we saw that South American continent had the most positive rate amongst other continent, so we would want to know which countries in south america had more positive rates than the continent's positive rates hence driving it upwards

```python
south_american_countries=covid_df[covid_df.Continent=="South America"]
south_american_countries.nunique()
```

```python
high_positive_rate_countries=south_american_countries[south_american_countries.positive_rate>south_american_countries.positive_rate.mean()].sort_values("positive_rate", ascending=False)
```

```python
sns.barplot(x=high_positive_rate_countries["Country/Region"],y=high_positive_rate_countries["positive_rate"])
plt.xticks(rotation=70);
```

#### A barplot was plotted to decipher the countries with higher positive rates than the continental death rate of South America and it shows that Suriname which was the country with the highest positive rates was one of them followed by the Bolivia and 2 other countries

#### Q3: TODO - What is the recovery rate per continent

```python
continent_recovery_rate=covid_df.groupby("Continent")["TotalRecovered"].sum()/covid_df.groupby("Continent")["TotalCases"].sum() *100 
sns.barplot(x=continent_recovery_rate, y=continent_recovery_rate.index);
```

```python
plt.pie(continent_recovery_rate, labels=continent_recovery_rate.index, autopct='%1.1f%%', startangle=180);
```

#### The recovery rate by continent was plotted and it showed that the recovery rate was highest in Asia and lowest in Europe and North America

#### Q4: TODO - From the recovery rate we see that although the recovery rates cutting across all continent is almost the same, Asia has the highest recovery rate and Europe and North America has the lowest recovery rate. Which countries has recovery rates higher than the continental recovery rate

```python
covid_df["recovery_rate"]= covid_df["TotalRecovered"]/covid_df["TotalCases"]*100 
highest_recovery_rate=covid_df.sort_values("recovery_rate",ascending=False).head(15)
sns.barplot(x="Country/Region", y="recovery_rate", data=highest_recovery_rate)
plt.xticks(rotation=70);
```

```python
lowest_recovery_rate=covid_df.sort_values("recovery_rate").head(15)
sns.barplot(x="Country/Region", y="recovery_rate", data=lowest_recovery_rate)
plt.xticks(rotation=70);
```

```python
asian_countries=covid_df[covid_df["Continent"]=="Asia"].sort_values('recovery_rate')
high_recovery_rate=asian_countries[asian_countries["recovery_rate"]>asian_countries.recovery_rate.mean()]
high_recovery_rate[["Country/Region","recovery_rate"]].sort_values("recovery_rate", ascending=False)
```

```python
sns.barplot(x=high_recovery_rate["Country/Region"],y=high_recovery_rate["recovery_rate"])
plt.xticks(rotation=70);
```

#### The recovery rate of countries with continent with the lowest recovery rate and highest recovery rate was plotted as seen above

#### Q5: TODO - Which country has the lowest recovery rate

```python
europe_countries=covid_df[covid_df["Continent"]=="Europe"].sort_values('recovery_rate')
lower_recovery_rate=europe_countries[europe_countries["recovery_rate"]<europe_countries.recovery_rate.mean()]
lower_recovery_rate[["Country/Region","recovery_rate"]].sort_values("recovery_rate")
sns.barplot(x=lower_recovery_rate["Country/Region"], y=lower_recovery_rate["recovery_rate"])
plt.xticks(rotation=70);
```

#### The countries with the lowest recovery rate and highest recovery rate irrespective of continent was plotted as seen above
